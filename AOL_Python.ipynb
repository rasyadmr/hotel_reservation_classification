## Feature Engineering

### Remove unwanted feature

```python
df_clean = df_raw.copy().drop('Booking_ID', axis = 1)
df_clean.reset_index(inplace = True, drop = True)
df_clean.head()
```

### Encoding

```python
df_num = df_clean.copy()

scale_Y = {"Not_Canceled" : 0, "Canceled" : 1}
scale_meal = {"Not Selected": 0, "Meal Plan 1": 1, "Meal Plan 2": 2, "Meal Plan 3": 3}
scale_room = {"Room_Type 1": 1, "Room_Type 2": 2, "Room_Type 3": 3, "Room_Type 4": 4, "Room_Type 5": 5, "Room_Type 6": 6, "Room_Type 7": 7}
scale_segment = {"Offline": 0, "Online": 1, "Corporate": 2, "Complementary": 4, "Aviation": 5}

df_encode = df_num[['booking_status', 'type_of_meal_plan', 'room_type_reserved', "market_segment_type"]]
df_encode['encode_Y'] = df_encode['booking_status'].replace(scale_Y)
df_encode['encode_meal'] = df_encode['type_of_meal_plan'].replace(scale_meal)
df_encode['encode_room'] = df_encode['room_type_reserved'].replace(scale_room)
df_encode['encode_segment'] = df_encode['market_segment_type'].replace(scale_segment)

temp = pd.concat([df_num, df_encode['encode_Y'], df_encode['encode_meal'], df_encode['encode_room'], df_encode['encode_segment']], axis = 1)
temp = temp.drop(['type_of_meal_plan', 'room_type_reserved', 'booking_status', 'market_segment_type'], axis = 1)
df_encoded = temp.copy()
```

### Feature addition

```python
df_add = df_encoded.copy()
df_add['total_people'] = df_add['no_of_adults'] + df_add['no_of_children']
df_add['total_night'] = df_add['no_of_week_nights'] + df_add['no_of_weekend_nights']
df_add['total_booking_history'] = df_add['no_of_previous_cancellations'] + df_add['no_of_previous_bookings_not_canceled']
df_add['spend'] = df_add['total_night'] * df_add['avg_price_per_room']
df_add['cancel_chance'] = np.where(df_add['total_booking_history'] < 1, df_add['total_booking_history'], df_add['no_of_previous_bookings_not_canceled'] / df_add['total_booking_history'])
df_add.drop(['no_of_adults', 'no_of_children', 'no_of_week_nights', 'no_of_weekend_nights', 'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled'], axis = 1, inplace = True)

df_feat = df_add.copy()
df_feat.head(20)
```

### Features Correlation

```python
df_corr = df_feat.copy().corr()
plt.figure(figsize = (16, 8))
sns.heatmap(df_corr, annot = True)
plt.show()
```

```python
df_corr = df_feat.copy().corr()[['encode_Y']].sort_values(by = 'encode_Y', ascending = True)
plt.figure(figsize = (4, 6))
sns.heatmap(df_corr, annot = True)
plt.show()
```

### Feature Selection

```python
threshold = 0.1
df_corr = df_feat.copy().corr()[['encode_Y']].sort_values(by = 'encode_Y', ascending = True)
col_selection = df_corr.where(abs(df_corr) >= threshold).T.dropna(axis = 1).columns

for i in col_selection:
    print(i)

df_final = df_feat.copy()[col_selection]
```

## Model

### Import Models & Utilities

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

### Splitting Train Test Data (Unengineered)

```python
X_train, X_val, y_train, y_val = train_test_split(df_feat.drop('encode_Y', axis = 1), df_feat['encode_Y'], test_size = 0.3)
```

### Logistic Regression

```python
log_reg = LogisticRegression(n_jobs = 4)
log_reg.fit(X_train, y_train)

predict = log_reg.predict(X_val)
report = classification_report(y_val, predict)
log_f1_un = f1_score(y_val, predict)
print(report)
```

### Random Forest Classifier

```python
rforest = RandomForestClassifier()
rforest.fit(X_train, y_train)

predict = rforest.predict(X_val)
report = classification_report(y_val, predict)
rf_f1_un = f1_score(y_val, predict)
print(report)
```

### Decision Tree Classifier

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)

predict = dtree.predict(X_val)
report = classification_report(y_val, predict)
dt_f1_un = f1_score(y_val, predict)
print(report)
```

### Splitting Train Test Data (Engineered)

```python
X_train, X_val, y_train, y_val = train_test_split(df_final.drop('encode_Y', axis = 1), df_final['encode_Y'], test_size = 0.3)
```

### Logistic Regression

```python
log_reg = LogisticRegression(n_jobs = 4)
log_reg.fit(X_train, y_train)

predict = log_reg.predict(X_val)
report = classification_report(y_val, predict)
log_f1_en = f1_score(y_val, predict)
print(report)
```

### Random Forest Classifier

```python
rforest = RandomForestClassifier()
rforest.fit(X_train, y_train)

predict = rforest.predict(X_val)
report = classification_report(y_val, predict)
rf_f1_en = f1_score(y_val, predict)
print(report)
```

### Decision Tree Classifier

```python
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)

predict = dtree.predict(X_val)
report = classification_report(y_val, predict)
dt_f1_en = f1_score(y_val, predict)
print(report)
```

### Comparison between Engineered and Unengineered

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (10, 3))

ax1.set_title('Logistic Regression F1-Score')
ax1.bar(['Unengineered', 'Engineered'], [log_f1_un, log_f1_en])
ax2.set_title('Random Forest F1-Score')
ax2.bar(['Unengineered', 'Engineered'], [rf_f1_un, rf_f1_en])
ax3.set_title('Decision Tree F1-Score')
ax3.bar(['Unengineered', 'Engineered'], [dt_f1_un, dt_f1_en])
```

There's a little difference of their F1-Score. We choose to use Unengineered data.

## Model Evaluate

### Import Utilities

```python
from sklearn.metrics import f1_score, roc_auc_score, plot_roc_curve
```

### Create Functions and Holder List

```python
def eval_model(model, X_test, y_test, X_val, y_val):
    model.fit(X_test, y_test)
    y_pred = model.predict(X_val).round()

    f1 = f1_score(y_val, y_pred)
    roc_score = roc_auc_score(y_val, y_pred)
    matrix = confusion_matrix(y_val, y_pred)

    return f1, roc_score, matrix

def roc_plot(model, X_test, y_test):
    ax = plt.subplot()
    plot_roc_curve(model, X_test, y_test, ax = ax)
    ax = plt.plot([0, 1], [0, 1], color="navy", linestyle="--")
    return ax

model_data = []
models = [log_reg, rforest, dtree]
```

### Storing Data into List

```python
model_data.clear()
for i in models:
    f1, roc_score, matrix = eval_model(i, X_train, y_train, X_val, y_val)
    model_data.append([type(i).__name__, f1, roc_score, matrix])
```

### Visualization of F1-Score

```python
for i in range(0, len(model_data)):
    plt.bar(model_data[i][0], model_data[i][1])
plt.title('F1 Scores')
plt.show()
```

### Visualization of ROC Score

```python
for i in range(0, len(model_data)):
    plt.bar(model_data[i][0], model_data[i][2])
plt.title('ROC Scores')
plt.show()
```

### Visualization of ROC Curve

```python
fig, ax = plt.subplots(1, 3, figsize = (15, 3))
for i in range(0, len(models)):
    plot_roc_curve(models[i], X_train, y_train, ax = ax[i])
    ax[i].plot([0, 1], [0, 1], color="navy", linestyle="--")
```

### Visualization of Confussion Matrix

```python
fig, ax = plt.subplots(1, 3, figsize = (15,3))

for i in range(0, len(model_data)):
    ax[i].set_title(model_data[i][0])
    sns.heatmap(model_data[i][3] / np.sum(model_data[i][3]), annot = True, cmap = 'Blues', fmt = '.2%', ax = ax[i])

plt.show()
```
